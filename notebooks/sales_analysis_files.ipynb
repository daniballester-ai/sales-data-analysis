## ADICIONAR A COLUNA ESTADO

```python
# criar uma coluna para o estado "state" que deve ser extraído da coluna "Purchase Address". O estado está após a cidade, são as duas letras após a segunda vírgula.

df['state'] = df['Purchase Address'].apply(lambda x: x.split(',')[2].strip()[:2])
```

```python
# verificar dados nulos
df.isnull().sum()
```

## ADICIONAR A REGIÃO DOS ESTADOS UNIDOS

```python
# criar uma coluna no dataframe df com a informação da região  dos Estados Unidos baseado na coluna state

def get_region(state):
  northeast = ['ME', 'NH', 'VT', 'MA', 'RI', 'CT', 'NY', 'NJ', 'PA']
  midwest = ['ND', 'SD', 'NE', 'KS', 'MN', 'IA', 'MO', 'WI', 'IL', 'MI', 'IN', 'OH']
  south = ['DE', 'MD', 'DC', 'VA', 'WV', 'NC', 'SC', 'GA', 'FL', 'KY', 'TN', 'AL', 'MS', 'AR', 'LA', 'OK', 'TX']
  west = ['WA', 'OR', 'CA', 'NV', 'ID', 'MT', 'WY', 'UT', 'CO', 'AZ', 'NM']

  if state in northeast:
    return 'Northeast'
  elif state in midwest:
    return 'Midwest'
  elif state in south:
    return 'South'
  elif state in west:
    return 'West'
  else:
    return 'Unknown'

df['Region'] = df['state'].apply(get_region)
```

```python
#exportar o dataframe df para um arquivo csv

df.to_csv('/content/drive/MyDrive/Python/df.csv', index=False)
```

```python
# dividir o arquivo df.csv em 20

n = 20  # Número de dataframes desejados
num_lines = len(df)
lines_per_df = math.ceil(num_lines / n)

for i in range(n):
  start = i * lines_per_df
  end = (i + 1) * lines_per_df
  df_subset = df.iloc[start:end]
  df_subset.to_csv(f'/content/drive/MyDrive/Python/vendas_{i + 1}.csv', index=False)
```
